```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
%%sql
create table if not EXISTS lh_ecormece.gold_fact_sale
(
    Order_ID	string	,
    Price float,
    Quantity	float	,
    Sales	float	,
    Discount	float	,
    Profit	float	,
    Shipping_Cost	float	,
    Order_Date	date	,
    Shipping_Date	date	,
    Product_ID	long	,
    OrderPriority_ID	Long	,
    ShipMode_ID	Long	,
    Customer_ID	string	,
    Order_Year	integer	,
    Order_Month	integer	,
    Created_TS	timestamp	,
    Modified_TS	timestamp	
)
using DELTA
PARTITIONED by (Order_Year,Order_Month)
```

```python
df_bronze= spark.sql("""select 
    bronze_sales.Order_ID	,
    bronze_sales.Sales as Price,
    bronze_sales.Quantity,
    bronze_sales.Sales	* bronze_sales.Quantity as Sales,
    bronze_sales.Discount	,
    bronze_sales.Profit	,
    bronze_sales.Shipping_Cost	,
    bronze_sales.Order_Date	,
    bronze_sales.Shipping_Date	,
    gold_dimproduct.Product_ID	,
    gold_dimorderpriority.OrderPriority_ID	,
    gold_dimshipmode.ShipMode_ID	,
    bronze_sales.Customer_ID	,
    Year(Order_Date) as Order_Year	,
    Month(Order_Date) as Order_Month	
from lh_ecormece.bronze_sales 
inner join lh_ecormece.gold_dimproduct on bronze_sales.Product=gold_dimproduct.Product and 
                                             bronze_sales.Product_Category=gold_dimproduct.Product_Category
inner join lh_ecormece.gold_dimshipmode on bronze_sales.Ship_Mode= gold_dimshipmode.Ship_Mode
inner join lh_ecormece.gold_dimorderpriority on bronze_sales.Order_Priority=gold_dimorderpriority.Order_Priority
""")
```

```python
df_bronze.createOrReplaceTempView("ViewFactSale")
```

```python
%%sql
merge into lh_ecormece.gold_fact_sale as gfs
    using ViewFactSale as vfs 
        on gfs.Order_Year=vfs.Order_Year and gfs.Order_Month=vfs.Order_Month and gfs.Order_ID=vfs.Order_ID
when matched then 
update SET
    gfs.Sales	=	vfs.Sales	,
    gfs.Price	=	vfs.Price	,
    gfs.Quantity	=	vfs.Quantity	,
    gfs.Discount	=	vfs.Discount	,
    gfs.Profit	=	vfs.Profit	,
    gfs.Shipping_Cost	=	vfs.Shipping_Cost	,
    gfs.Order_Date	=	vfs.Order_Date	,
    gfs.Shipping_Date	=	vfs.Shipping_Date	,
    gfs.Product_ID	=	vfs.Product_ID	,
    gfs.OrderPriority_ID	=	vfs.OrderPriority_ID	,
    gfs.ShipMode_ID	=	vfs.ShipMode_ID	,
    gfs.Customer_ID	=	vfs.Customer_ID	,
    gfs.Modified_TS	=	current_timestamp()	

when not matched then
INSERT
(
    gfs.Order_ID	, --1
    gfs.Sales	,
    gfs.Price,
    gfs.Quantity	,
    gfs.Discount	,
    gfs.Profit	,
    gfs.Shipping_Cost	,
    gfs.Order_Date	,
    gfs.Shipping_Date	,
    gfs.Product_ID	,
    gfs.OrderPriority_ID	,
    gfs.ShipMode_ID	,
    gfs.Customer_ID	,
    gfs.Order_Year	, --2
    gfs.Order_Month	, --3
    gfs.Created_TS	,
    gfs.Modified_TS	

)
VALUES
(
    vfs.Order_ID	,
    vfs.Sales	,
    vfs.Price,
    vfs.Quantity	,
    vfs.Discount	,
    vfs.Profit	,
    vfs.Shipping_Cost	,
    vfs.Order_Date	,
    vfs.Shipping_Date	,
    vfs.Product_ID	,
    vfs.OrderPriority_ID	,
    vfs.ShipMode_ID	,
    vfs.Customer_ID	,
    vfs.Order_Year	,
    vfs.Order_Month	,
    current_timestamp()	,
    current_timestamp()
)
```
